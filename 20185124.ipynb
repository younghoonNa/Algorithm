````r
Kim <- Reg_DF %>% filter(batter_name == "김강민")
Kim
````

````r
# 결측값 처리
# 1. 2002 년도 기록에 NA 값이 포함되어있다. -> 2012년도에 1게임에 나와서 타석x, 안타x 이기 때문에 타격 지표인 타율 avg , 장타율 Slg, 출루율 OBP가 na로 되어있다. 출루율 + 장타율인 OPS는 장타율과 출루율이 NA 이므로
# 똑같이 NA 값을 나타내고 있다.

# 결측값이 존재하는 속성만 제외하고 평균 구하기 vs 결측값을 포함한 다른 지표도 같이 없애버리기
# 2002년도에는 타격 지표가 1게임 출장밖에 없으므로 결측값을 포함한 다른 지표도 제거하는 것이 더 좋을 것 같다고 생각을 하였습니다.

#na.omit를 통해 NA 값이 있는 2002 년도 레코드제거

Kim = na.omit(Kim)
Kim
````

````r
Kim_graph <- Kim %>% select(batter_name,year, G, H, R)
head(Kim_graph)
Kim_graph_3 = rbind(Kim_graph, Kim_graph, Kim_graph) 
Kim_graph_3<- Kim_graph_3 %>% arrange((year))
Kim_graph_3


````

````r
# vec = rep(c("G", "H", "R"), times = 16)
# Kim_graph_3 = cbind(Kim_graph_3, vec)
# Kim_graph_3

fac = c()
val = c()
year_c = c()

year = 2003
i = 1 
j = 1


while(year <= 2018) {
  for(k in 1:3) {
    if (k%%3 == 1) {
      fac[i] = "G"
      val[i] = Kim_graph[j,3]
      year_c[i] = year
    } else if (k%%3 == 2) {
      fac[i] = "H"
      val[i] = Kim_graph[j,4]
      year_c[i] = year
    } else {
      fac[i] = "R"
      val[i] = Kim_graph[j,5]
      year_c[i] = year
    }
    i = i + 1
  }
  year = year +1
  j = j + 1

}

fac = factor(fac)
fac
val
year_c
````

````r
i= 1
temp = c()
for(i in 1:48) {
  temp[i] = val[[i]]
}
print(temp)
val = temp
str(val)
````

````r
Kim_graphgg = cbind(Kim_graph_3, fac, val)
Kim_graphgg <- Kim_graphgg[,-3] 
Kim_graphgg <- Kim_graphgg[,-3]
Kim_graphgg <- Kim_graphgg[,-3]
Kim_graphgg
````

````r
Kim_graphgg %>% ggplot(aes(year, val, col = fac))+ geom_point() + geom_line()
# Kim_graph %>% ggplot(aes(year, R, col = "0000FF"))+ geom_point()+ geom_line()
#  scale_x_continuous(breaks = seq(2002, 2018, 1)) +
#   scale_y_continuous(breaks = seq(0, 350, 50)) +


````

````r
year = 2003
i = 1
j = 1
l = 1

year_tmp = c()
temp_fac = c()
num  = c()

while(year <= 2018) {
    

    data = Kim_graphgg[[j,2]]

    if(data != year){
      year = year + 1
    } 


    k = Kim_graphgg[j,4]

    # print(k)
    while(i<=k){
      year_tmp[l] = year
      temp_fac[l] = Kim_graphgg[j,3]     
      num[l] = i 
      i = i + 1
      l = l + 1
      # print(k)
      # print(Kim_graphgg[j,3])
    }
    
    j = j + 1
    i = 1
    

    if(j > 48)
      break
     
}

t = 1
str = c()
int_year = c()

int_year <- as.integer(year_tmp)
num <- as.integer(num)

for(t in 1:3163) {
  str[t] = temp_fac[[t]]

  if(str[t] == 1) {
    str[t] = "G"
  } else if(str[t] == 2) {
    str[t] = "H"
  } else {
    str[t] = "R"
  }
}

str <- as.factor(str)
````

````r

Kim_geom_hist = data.frame(int_year, str, num)
head(Kim_geom_hist)
````
